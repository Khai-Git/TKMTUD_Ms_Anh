### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**. Tính:

a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sử dụng hàm seed() dùng để tái lập thí nghiệm. Bạn có thể truyền tham số bất kỳ
np.random.seed(26)
```

```python
N = 10000  #kích thước quần thể
POP = np.random.randint(120, 200, N)
```

```python
# a.b. Tỷ lệ người cao
tyle_pop = (POP>=180).sum() / N
trungbinh_pop = POP.mean()
dlc_pop = POP.std()

print('Tham số của Quần Thể:')
print('- Tỷ  lệ người cao: {} \n- Kỳ vọng: {} \n- Độ lệch chuẩn: {}'.format(tyle_pop, trungbinh_pop, dlc_pop.round(4)))
```

```python
# c. Histogram
plt.hist(POP, edgecolor='white', bins=30)
# plt.hist(POP, edgecolor='white')
plt.show()
```

```python
from scipy.stats import norm
```

```python
# d.Vẽ phân phối chuẩn
x = np.linspace(trungbinh_pop - 4*dlc_pop, trungbinh_pop + 4*dlc_pop, 100)
y = norm.pdf(x, loc=trungbinh_pop, scale=dlc_pop)
plt.plot(x, y)
plt.title('Phân phối chuẩn $\mu={}, \sigma={}$'.format(trungbinh_pop, dlc_pop.round(4)))
plt.show()
```

**<u>Nhận xét</u>** 

Có thể nhận thấy đồ thị histogram của quần thể không có dạng phân phối chuẩn (hình dáng khác với đồ thị hàm mật độ phân phối chuẩn ta vẽ mô phỏng), vì vậy ta có thể kết luận quần thể ban đầu không có phân phối chuẩn.

### Bài 2: Lấy mẫu cỡ 20

Bạn hãy lấy một mẫu kích thước 20 phần từ. Tính:

a. Trung bình mẫu, và độ lệch chuẩn của mẫu

b. Tỷ lệ người cao của mẫu

c. Vẽ histogram của mẫu

d. Thử vẽ đồ thị hàm mật độ của phân phối chuẩn với tham số loc và scale bằng với giá trị trung bình và độ lệch chuẩn và chiều cao của mẫu. Bạn có nhận xét gì không?

```python
# a. Trung bình, độ lệch chuẩn của mẫu
sample_size = 20
sam = np.random.choice(POP, size=sample_size)
tb_mau = sam.mean()
dlc_mau = sam.std()
print('Mẫu kích thước 20: ', sam)
print('Trung bình mẫu: ', tb_mau)
print('Độ lệch chuẩn: ', dlc_mau.round(4))

# b. 
print('Tỷ lệ người cao: ', (sam>=180).sum()/sample_size)

# c. Historam
plt.hist(sam, edgecolor='w', bins=20)
plt.show()

# d. Phân phối chuẩn
x = np.linspace(tb_mau - 4*dlc_mau, tb_mau + 4*dlc_mau, 100)
y = norm.pdf(x, loc=tb_mau, scale=dlc_mau)
plt.plot(x, y)
plt.title('Phân phối chuẩn $\mu={}, \sigma={}$'.format(tb_mau, dlc_mau.round(4)))
plt.show()
```

**<u>Nhận xét</u>** 

Tương tự câu 1, ta cũng có thể thấy phân phối chiều cao các thanh niên trong mẫu không có phân phối chuẩn.

### Bài 4: Khảo sát phân phối của trung bình mẫu

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

#### a. Mô phỏng trung bình mẫu cỡ 20

```python
# a) Lấy mẫu ngẫu nhiên (n = 20) 100 lần. Vẽ histogram cho 100 trung bình mẫu thu được
arr_trungbinh_20 = []
for i in range(100):
    trungbinhmau = np.random.choice(POP, size=20).mean()
    arr_trungbinh_20.append(trungbinhmau)

plt.hist(arr_trungbinh_20, edgecolor='w', bins=25)
plt.title('Histogram của trung bình mẫu cỡ 20')
plt.xlabel('Chiều cao trung bình mẫu')
plt.ylabel('Tần số')
plt.show()
```

#### b. Mô phỏng trung bình mẫu cỡ 100

```python
# b) Lấy mẫu ngẫu nhiên (n = 100) 100 lần. Vẽ histogram cho 100 trung bình mẫu thu được
arr_trungbinh_100 = []
for i in range(100):
    trungbinhmau = np.random.choice(POP, size=100).mean()
    arr_trungbinh_100.append(trungbinhmau)

plt.hist(arr_trungbinh_100, edgecolor='w', bins=25)
plt.title('Histogram của trung bình mẫu cỡ 100')
plt.xlabel('Chiều cao')
plt.ylabel('Tần số')
plt.show()
```

#### c. So sánh hai phân phối

```python
# c) Vẽ histogram của hai câu a, b trong cùng một hình để so sánh
plt.hist(arr_trungbinh_20, edgecolor='w', bins=25, label='cỡ mẫu: 20')
plt.hist(arr_trungbinh_100, edgecolor='w', bins=25, label='cỡ  mẫu: 100')
plt.legend()
plt.xlabel('Chiều cao')
plt.ylabel('Tần số')
plt.show()
```

**<u>Nhận xét</u>** 

Đồ thị histogram mô phỏng hai phân phối của trung bình mẫu với hai cỡ mẫu khác nhau là 20 và 100 có đặc điểm:

- Cả hai đều có hình dạng gần giống như phân phối chuẩn, với vị trí của chiều cao trung bình là như nhau.

- Histogram của đồ thị cỡ mẫu 100 có phần đuôi hẹp hơn, nghĩa là độ lệch ít hơn so với phân phối mẫu có cỡ là 20. Điều này có nghĩa là nếu ta lấy mẫu kích thước lớn hơn thì sai số của ước lượng sẽ nhiều khả năng nhỏ hơn.

### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


#### a. Biết $\sigma = 0.3$ 

##### Cách 1 (tính tay)

Từ mẫu đã cho ta tính được: 

$ \begin{cases} n = 9 \\ \bar{x} = 1.9333 \\ s = 0.255 \end{cases}$

a, Biết $\sigma = 0.3$

Với độ tin cậy là 95%, ta có: $z_{\alpha / 2} = 1.96$. (Với $\phi(z_{\alpha / 2}) = \frac{1}{2} - \frac{\alpha}{2} = \frac{1-\alpha}{2})$ - Tra bảng phân phối z)

Sai số ước lượng là:
$$\epsilon = z_{\alpha / 2}\frac{\sigma}{\sqrt{n}} = 1.96\frac{0.3}{\sqrt{9}} = 0.196$$

Vậy khoảng ước lượng cho $\mu$ là:

$$(1.9333 - 0.196; 1.9333 + 0.196) = (1.7373; 2.1293)$$

##### Cách 2 (dùng python)

```python
data_ga = [2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8]

sigma = 0.3

n = len(data_ga)
tb = np.mean(data_ga)
dlc = np.std(data_ga, ddof=1)

print('Các thống kê mẫu: ')
print('Số lượng: ', n)
print('Trung bình mẫu: ', tb.round(4))
print('Độ lệch chuẩn mẫu: ', dlc.round(4))
```

```python
dotincay = 0.95
alpha_chia_2 = np.float64((1-dotincay)/2)
print('alpha/2: ', alpha_chia_2.round(4))
```

```python
import math

# a. Biết sigma = 0.3 --> Phân phối chuẩn tắc (z)
saiso = (norm.isf(alpha_chia_2) * sigma) / math.sqrt(n)
print('Sai số ước lượng: ', saiso.round(4))
print('Khoảng ước lượng cho trung bình: [{}, {}]'.format((tb-saiso).round(4), (tb+saiso).round(4)))
```

##### Cách 3 (dùng thư viện)

Dựa vào việc xác định trung bình mẫu có phân phối chuẩn (dựa vào một số dữ kiện của đề bài). Ta có thể sử dụng hàm `interval()` từ thư viện scipy.stats

```python
dlc_trungbinh = np.float64(sigma/math.sqrt(n))
print('Khoảng ướng lượng: ')
norm.interval(0.95, loc=tb, scale=dlc_trungbinh)
```

##### Minh họa

Giả sử ta biết được đại lượng thống kê trung bình mẫu có phân phối chuẩn, với trung bình và độ lệch chuẩn xác định. Ta có thể dể dàng tính được cận trên và cận dưới của khoảng ước lượng. Bạn có thể quan sát hình vẽ dưới đây:

```python
# Định nghĩa hàm vẽ hàm mật độ phân phối chuẩn để minh hoạ
def drawNormalHist(dotincay, kyvong, dolechchuan):
    canduoi_2 = kyvong - 4 * dolechchuan
    canduoi_1 = norm.ppf((1 - dotincay) / 2, loc=kyvong, scale=dolechchuan)
    cantren_1 = norm.isf((1 - dotincay) / 2, loc=kyvong, scale=dolechchuan)
    cantren_2 = kyvong + 4*dolechchuan
           
    plt.fill_between(x=np.arange(canduoi_2, canduoi_1, 0.01), 
                     y1= norm.pdf(np.arange(canduoi_2, canduoi_1, 0.01), loc=kyvong, scale=dolechchuan),
                     color='red',
                     alpha=0.35)

    plt.fill_between(x=np.arange(cantren_1, cantren_2, 0.01), 
                     y1= norm.pdf(np.arange(cantren_1, cantren_2, 0.01), loc=kyvong, scale=dolechchuan),
                     color='red',
                     alpha=0.35)

    x = np.arange(canduoi_1, cantren_1, 0.01)
    y = norm.pdf(x, loc=kyvong, scale=dolechchuan)
    plt.plot(x, y)

    plt.fill_between(x, y, color='b', alpha=0.35)
    plt.title('Phân phối chuẩn $\mu={}, \sigma={}$'.format(kyvong.round(4), dolechchuan.round(4)))
    
    plt.text(x=1.92, y=0.5, s= round(dotincay,3), size=15)
    plt.text(x=2.2, y=0.3, s= round((1-dotincay)/2, 3), size=12)
    plt.text(x=1.64, y=0.3, s= round((1-dotincay)/2, 3), size=12)
    
    print('Khoảng tin cậy: [{}, {}]'.format(canduoi_1.round(4), cantren_1.round(4)))
    
    plt.show()
```

```python
plt.subplots(figsize=(15, 6))

dlc_trungbinh = np.float64(sigma/math.sqrt(n))

drawNormalHist(0.95, tb, dlc_trungbinh)
```

```python
# Tính cận trên
norm.isf(0.025, loc=tb, scale=dlc_trungbinh).round(4)
```

```python
# Tính cận dưới (dùng một trong hai câu lệnh dưới đây)
norm.ppf(0.025, loc=tb, scale=dlc_trungbinh).round(4)

# norm.isf(0.975, loc=tb, scale=dlc_trungbinh).round(4)
```

##### Nhận xét

Nếu biết đại lượng cần ước lượng có phân phối chuẩn, bạn có thể sử dụng các hàm hỗ trợ của python để tính trực tiếp mà không cần phải chuyển về phân phối chuẩn tắc nữa.
