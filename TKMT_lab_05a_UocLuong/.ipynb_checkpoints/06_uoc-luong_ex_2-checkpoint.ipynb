## <u>Bài 1</u>

Green M&M Candies liên quan đến Dataset 18 trong file excel. Tìm tỉ lệ mẫu của M&M có màu xanh lá cây (green). Sử dụng kết quả để xây dựng 1 ước lượng khoảng tin cậy 95% của % quần thể M&M có màu xanh.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('./data/18_M_M.xls', index_col=0)
df
```

```python
# Tìm số lượng M&M có màu xanh lá cây
num_green = df['Green'].count()
print(num_green)
# Tìm kích thước mẫu
n = len(df)
print(n)
# Tỉ lệ mẫu của M&M có màu xanh lá cây
p_hat = num_green / n
print(p_hat)
# Ước lượng khoảng tin cậy 95%
z = 1.96
se = np.sqrt(p_hat * (1 - p_hat) / n)
lower_bound = p_hat - z * se
upper_bound = p_hat + z * se

print('Tỉ lệ mẫu của M&M có màu xanh lá cây:', p_hat)
print('Ước lượng khoảng tin cậy 95%:', lower_bound, upper_bound)
```

## <u>Bài 2</u>

Freshman Weight Gain liên quan đến Dataset 3 trong file excel

a. Dựa vào kết quả của mẫu, tìm ước lượng điểm tốt nhất của tỉ lệ phần trăm các sinh viên cao đẳng tăng cân trong năm thứ 1.

b. Xây dựng ước lượng khoảng tin cậy 95% về tỉ lệ phần trăm các sinh viên cao đẳng tăng cân trong năm thứ 1.


```python
import numpy as np
import pandas as pd

# Đọc dữ liệu từ tệp Excel
df = pd.read_excel('./data/03_FRESH15.xls', index_col=0)
df
```

```python
# Tìm số lượng sinh viên tăng cân
num_WTSEP = df['WTSEP'].count()

# Tìm kích thước mẫu
n = len(df)

# Ước lượng điểm tốt nhất
p_hat = num_WTSEP / n

# Ước lượng khoảng tin cậy 95%
z = 1.96
se = np.sqrt(p_hat * (1 - p_hat) / n)
lower_bound = p_hat - z * se
upper_bound = p_hat + z * se

print('Ước lượng điểm tốt nhất:', p_hat)
print('Ước lượng khoảng tin cậy 95%:', lower_bound, upper_bound)
```

## <u>Bài 3</u>

Lượng mưa ở Boston: liên quan đến Dataset 14 trong file excel, và quan tâm đến các ngày với các giá trị lượng mưa khác nhau từ 0 đến các ngày có mưa có giá trị lượng mưa lớn hơn 0. Xây dựng ước lượng khoảng tin cậy 95% cho tỉ lệ mưa trong các ngày Thứ Tư và xây dựng ước lượng khoảng tin cậy 95% cho tỉ lệ mưa trong các ngày Chủ Nhật.

```python
import numpy as np
import pandas as pd

# Đọc dữ liệu từ tệp Excel
df = pd.read_excel('./data/14_BOSTRAIN.xls', index_col=0)
df
```

```python
# Chỉ giữ các ngày có lượng mưa khác 0
df = df > 0

# Tìm số lượng ngày mưa vào Thứ Tư
num_rain_wednesday = df['WED'].count()

# Tìm số lượng ngày mưa vào Chủ Nhật
num_rain_sunday = df['SUN'].count()

# Tìm kích thước mẫu
n_wednesday = len(df['WED'])
n_sunday = len(df['SUN'])

# Ước lượng điểm tốt nhất
p_hat_wednesday = num_rain_wednesday / n_wednesday
p_hat_sunday = num_rain_sunday / n_sunday

# Ước lượng khoảng tin cậy 95%
z = 1.96
se_wednesday = np.sqrt(p_hat_wednesday * (1 - p_hat_wednesday) / n_wednesday)
se_sunday = np.sqrt(p_hat_sunday * (1 - p_hat_sunday) / n_sunday)
lower_bound_wednesday = p_hat_wednesday - z * se_wednesday
upper_bound_wednesday = p_hat_wednesday + z * se_wednesday
lower_bound_sunday = p_hat_sunday - z * se_sunday
upper_bound_sunday = p_hat_sunday + z * se_sunday

print('Ước lượng khoảng tin cậy 95% cho tỉ lệ mưa trong các ngày Thứ Tư:', lower_bound_wednesday, upper_bound_wednesday)
print('Ước lượng khoảng tin cậy 95% cho tỉ lệ mưa trong các ngày Chủ Nhật:', lower_bound_sunday, upper_bound_sunday)
```

## <u>Bài 4</u>

Tổng số tiền phim: liên quan đến Dataset 9 trong file excel. Xây dựng ước tính khoảng thời gian tin cậy 95% của tổng số tiền trung bình cho quần thể của tất cả các phim. Giả định rằng độ lệch chuẩn của quần thể được biết là 100 triệu đô la. 

```python
# Đọc dữ liệu từ tệp Excel
df = pd.read_excel('./data/09_MOVIES.xls', index_col=0)
df
```

```python
# Tìm kích thước mẫu
n = len(df)

# Ước lượng điểm tốt nhất
mu_hat = df['Budget'].mean()

# Ước lượng khoảng tin cậy 95%
z = 1.96
se = 100 / np.sqrt(n)
lower_bound = mu_hat - z * se
upper_bound = mu_hat + z * se

print('Ước lượng khoảng tin cậy 95% của tổng số tiền trung bình cho quần thể của tất cả các phim:', 
      lower_bound, ' | ', upper_bound)
```
